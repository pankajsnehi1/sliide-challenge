## Using 'ga_session_number' and 'subscribed' as valuable events to determine the number of active users

```python
dates = []
for index, row in user_eng.iterrows():
    event_timestamp = user_eng['event_timestamp'].loc[index]
    event_time = time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime(event_timestamp/1000000))
    print('Event time-----', event_time)
    user_prop = user_eng['user_properties'][index]
    for event in user_prop:
        for k, v in event.items():
            if event['key'] == 'ga_session_number' or event['key'].startswith('subscribed') :
                for i, j in event['value'].items():
                    if i == 'set_timestamp_micros':
                        user_time = time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime(int(j)/1000000))
                        print('User time', user_time)
                        time_diff = time.mktime(time.gmtime(event_timestamp/1000000)) - time.mktime(time.gmtime(int(j)/1000000))
                        if time_diff > 3:
                            dates.append(user_time)
                            print('Time diff = ', time_diff)
```

## Preapring the dates dataframe with active users

```python
dates_dict = dict(Counter(dates))
dates_dict = {k[:10]:v for k, v in dates_dict.items()}
df = pd.DataFrame.from_dict([dates_dict]).T
df = df.reset_index()
df.rename(columns={'index':'date', 0:'active_users'})
```

## Creating a database to write the dataframe into

```python
conn = sqlite3.connect('sliide_challenge.db')
c = conn.cursor()
c.execute('CREATE TABLE active_user_table (date, active_user_count)')
conn.commit()
```

## Writing the dataframe into the database and confirming it's been written okay

```python
df.to_sql('active_user_table', conn, if_exists='replace', index = False)
c.execute('''select * from active_user_table''')
for row in c.fetchall():
    print (row)
```
